```python
#PCA mathematically computed over raw data
#author: Sourabh garg
from pandas import read_excel
import pandas as plt
import numpy as np
import matplotlib.pyplot as plt
load_data=read_excel("Data_files/data.xlsx",header=[0],sep=',')
load_num_data=load_data.as_matrix()
load_num_data
```

```python
mean_CdS=np.mean(load_data["CdS"])
mean_MoO=np.mean(load_data["MoO"])
mean_SnO2=np.mean(load_data["SnO2"])
mean_ZnO=np.mean(load_data["ZnO"])
mean_vector = np.array([[mean_CdS],[mean_MoO],[mean_SnO2],[mean_ZnO]])
print(mean_CdS,mean_MoO,mean_SnO2,mean_ZnO)
print(mean_vector.shape)
```

```python
std_CdS=np.std(load_data["CdS"])
std_MoO=np.std(load_data["MoO"])
std_SnO2=np.std(load_data["SnO2"])
std_ZnO=np.std(load_data["ZnO"])
print(std_CdS,std_MoO,std_SnO2,std_ZnO)
```

```python
#scatter matrix
scatter_matrix=np.zeros((4,4))
for i in range(load_data.shape[0]):
    scatter_matrix+=(load_num_data[i,:].reshape(4,1)-mean_vector).dot((load_num_data[i,:].reshape(4,1)-mean_vector).T)
print(scatter_matrix)
```

```python
cov_mat = np.cov([load_num_data[:,0],load_num_data[:,1],load_num_data[:,2],load_num_data[:,3]])
cov_mat
```

```python
eig_val_sc, eig_vec_sc = np.linalg.eig(scatter_matrix)
eig_val_cov, eig_vec_cov = np.linalg.eig(cov_mat)
print(eig_val_cov)
print("_________________________")
print(eig_vec_cov)
```

```python
eig_pairs = [(np.abs(eig_val_cov[i]), eig_vec_cov[:,i]) for i in range(len(eig_val_cov))]
eig_pairs.sort(key=lambda x: x[0], reverse=True)
feature_matrix=np.empty((0,4),float)

for i in eig_pairs:
    feature_matrix = np.append(feature_matrix, np.array([i[1]]), axis=0)
    #print(i[0], i[1])
print(feature_matrix)
```

```python
Mean_transformed_X=np.empty((0,42),float)
for i in range(4):
    Mean_transformed_X = np.append(Mean_transformed_X, np.array([load_num_data[:,i]-mean_vector[i]]), axis=0)

Mean_transformed_X=Mean_transformed_X.T
print(Mean_transformed_X)
```

```python
#print(feature_matrix)
#print(Mean_transformed_X)
PCA_transformed_data=Mean_transformed_X.dot(feature_matrix.T)
print(PCA_transformed_data)
#load_num_data.dot(FV)
print(type(PCA_transformed_data))
```
