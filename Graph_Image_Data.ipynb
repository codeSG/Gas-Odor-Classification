```python
from pandas import DataFrame
def convertt(x_c,y_c,x0,y0,x_max,y_max,x_range,y_range):
    conc=[]
    resp=[]
    for xi,yi in zip(x_c,y_c):
        xit,yit=xi-x0,yi-y0
        conc_xi=xit/((x_max-x0)/(x_range))
        resp_yi=yit/((y_max-y0)/(y_range))
        conc.append(conc_xi)
        resp.append(resp_yi)
    return conc,resp
```

```python
conc_z,resp_z=convertt(ax_ZnO,ay_ZnO,ax0,ay0,ax_max,ay_max,ax_range,ay_range)
conc_m,resp_m=convertt(ax_MoO,ay_MoO,ax0,ay0,ax_max,ay_max,ax_range,ay_range)
conc_c,resp_c=convertt(ax_CdS,ay_CdS,ax0,ay0,ax_max,ay_max,ax_range,ay_range)
conc_s,resp_s=convertt(ax_SnO2,ay_SnO2,ax0,ay0,ax_max,ay_max,ax_range,ay_range)
acetone=DataFrame()
acetone["Conc_Zn"]=conc_z
acetone["resp_Zn"]=resp_z
acetone["Conc_Cd"]=conc_c
acetone["resp_Cd"]=resp_c
acetone["Conc_Mo"]=conc_m
acetone["resp_Mo"]=resp_m
acetone["Conc_Sn"]=conc_s
acetone["resp_Sn"]=resp_s
acetone
```
